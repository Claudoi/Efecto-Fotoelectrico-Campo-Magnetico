
# Resolución del Problema: Movimiento de una Partícula Cargada en un Campo Magnético

**Problema**: Un electrón con una capacidad de trabajo (energía cinética) de \( 82090 	imes 10^{-19} \, 	ext{J} \) orbita de manera perpendicular a un campo magnético de \( 4250 \, 	ext{G} \).

1. ¿Cuál es el radio de la órbita?  
2. ¿Cuál es su frecuencia angular y periodo?  
3. Resolver el mismo problema para un antineutrón.

Representaremos gráficamente las trayectorias calculadas.


```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes físicas
carga_electron = -1.602e-19  # Carga del electrón (C)
masa_electron = 9.109e-31     # Masa del electrón (kg)
campo_magnetico = 4250e-4      # Campo magnético (T)
cinetica_electron = 82090e-19 # Energía cinética del electrón (J)
masa_antineutron = 1.675e-27  # Masa del antineutrón (kg)
```

```python
# Velocidad del electrón
v_electron = np.sqrt(2 * cinetica_electron / masa_electron)

# Radio de la órbita del electrón
r_electron = (masa_electron * v_electron) / (abs(carga_electron) * campo_magnetico)

# Frecuencia angular y periodo del electrón
w_electron = v_electron / r_electron
T_electron = 2 * np.pi / w_electron

# Resultados para el electrón
print(f"Electrón:")
print(f"Radio de la órbita: {r_electron:.2e} m")
print(f"Frecuencia angular: {w_electron:.2e} rad/s")
print(f"Periodo: {T_electron:.2e} s")
```

```python
# Velocidad del antineutrón (misma energía cinética)
v_antineutron = np.sqrt(2 * cinetica_electron / masa_antineutron)

# Radio de la órbita y otros parámetros para el antineutrón
# Antineutrón no tiene carga -> trayectoria no circular
r_antineutron = None
w_antineutron = None
T_antineutron = None

print(f"Antineutrón:")
print("Radio de la órbita: No aplica (sin carga, por tanto, trayectoria recta)")
print("Frecuencia angular: No aplica")
print("Periodo: No aplica")
```

```python
# Representación de la trayectoria del electrón
t = np.linspace(0, T_electron, 1000)  # Tiempo para una órbita
x_electron = r_electron * np.cos(w_electron * t)
y_electron = r_electron * np.sin(w_electron * t)

# Gráfica
plt.figure(figsize=(8, 8))
plt.plot(x_electron, y_electron, label="Electrón (Trayectoria circular)")
plt.title("Trayectoria del Electrón en un Campo Magnético")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.grid()
plt.show()
```
